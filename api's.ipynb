```python
tell_a_joke = [{'role':'user','content':'tell me a joke with emoji'}]
```

```python
from litellm import completion
```

```python
response = completion(model='gemini/gemini-2.5-flash',messages=tell_a_joke)
reply = response.choices[0].message.content
display(Markdown(reply))
```

```python
print(f'Input token: {response.usage.prompt_tokens}')
print(f'Output token: {response.usage.completion_tokens}')
print(f'Total tokens: {response.usage.total_tokens}')
print(f'Total cost: {response._hidden_params['response_cost']*100:.4f}')
```

```python
with open('hamlet.txt', 'r',encoding='utf-8') as f:
    hamlet = f.read()
loc = hamlet.find("Speak, man")
print(hamlet[loc:loc+100])
```

```python
question = [{'role':'user','content':"In Hamlet, when Laertes asks 'Where is my father?' what is the reply?"}]
```

```python
response = completion(model='gemini/gemini-2.5-flash',messages=question)
display(Markdown(response.choices[0].message.content))
```

```python
print(f'Input token:{response.usage.prompt_tokens}')
print(f'Ouput token: {response.usage.completion_tokens}')
print(f'Total token: {response.usage.total_tokens}')
print(f'Total cost: {response._hidden_params['response_cost'] * 100:.4f}')
```

```python
question[0]['content']+= '\n\n For context, Here is the entire text of Hamlet:\n\n'+hamlet
```

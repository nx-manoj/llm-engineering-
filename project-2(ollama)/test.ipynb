```python
import os 
from openai import OpenAI
from IPython.display import display, Markdown
```

```python
client = OpenAI(base_url="http://localhost:11434/v1/", api_key='ollama')
```

```python
system_prompt = "You are a helpful assistant. where you will summarize the data provided from website and provide the summary in a concise manner. You will also provide the key points from"

user_prompt_prefix = '''Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.'''
```

```python
from scrap import fetch_website_content
```

```python
def summarize_website(url):
    website_content = fetch_website_content(url)
    if not website_content:
        return "Failed to fetch website content."
    user_prompt =  user_prompt_prefix + website_content

    response = client.chat.completions.create(
        model="llama3.2:latest",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )
    summary = response.choices[0].message.content
    return summary
```

```python
def display_summary(url):
    summary = summarize_website(url)
    display(Markdown(f"### Summary of {url}\n\n{summary}"))
    
```

```python
display_summary("https://time.is/")
```

```python
display_summary("https://www.thehindu.com/")
```
